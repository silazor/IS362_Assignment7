```python
import pandas as pd
import os
os.chdir('C:/Users/silaz/github/capstone/is362/w7')
movie_ratings = pd.read_csv('reviews.csv')

movie_ratings.columns.str.strip().str.lower().str.replace(' ', '_')
movie_ratings['mean'] = movie_ratings.mean(axis=1)
movie_ratings
```

```python
movie_ratings.mean(axis=0, skipna=True).to_frame()
```

```python
movie_ratings_norm = movie_ratings.iloc[:,1:]
movie_ratings_norm.iloc[:,0:] = (movie_ratings_norm - movie_ratings_norm.mean()) / (movie_ratings_norm.max() - movie_ratings_norm.min())
movie_ratings_norm.columns = movie_ratings_norm.columns.str.strip().str.lower().str.replace(' ', '_')
movie_ratings_norm['mean'] = movie_ratings_norm.mean(axis=1)
movie_ratings_norm
```

```python
movie_ratings_norm.mean(axis=0, skipna=True).to_frame()
```

### Advantages and Disadvantages of Normalizing Data
An advantage of data normalization is the ability to change values of numeric columns in a dataset to a common scale. This helps to view the numeric values across the columns without distorting the numbers or ranges.

A disadvantage is that it doesn't work for every dataset and the outcome of normalizing data that doesn't require normalization can be unclear and confusing to read.

In this exercise, it seems that this dataset has a consistent range and provides inconsistent results when normalized.
